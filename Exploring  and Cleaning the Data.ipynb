Implementing OHE for a categorical  feature

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi']=400
df_clean_1=pd.read_csv('C:\\Users\\USER\\Desktop\\ESSY\\scores.csv')
```

```python
df_clean_1['ML_cat']='None'
```

```python
df_clean_1[['ML','ML_cat']].head(10)
```

```python
cat_mapping={
    1:'gradute school',
    0.63:'university',
    0.60:'high school',

    0.62:'college'
    
}
```

```python
df_clean_1['ML_cat']=df_clean_1['ML'].map(cat_mapping)
df_clean_1[['ML','ML_cat']].head(5)
```

```python
df_clean_1.ohe=pd.get_dummies(df_clean_1['ML_cat'])
df_clean_1.ohe.head()
```
